## Las Variables
1. NUM --> Valor a determinar signo
2. R --> Resultado del signo del valor

## Las Formulas
$NUM >= 0$

## Diagrama de flujo lucidchart
<img src="https://raw.githubusercontent.com/DanielGallegos128/Act-12-p3.2-Positivo-o-Negativo-/refs/heads/main/DF%20Positivo%20Negativo.png" width="300" height="500">

```python
# Para cuando es verdadero (si)
# Determinando si un numero es positivo o negativo usando if else
# Datos de entrada NUM , R
# Mostrar R = "Negativo" si es falso
# Si es verdadero mostrar R = "Positivo"
NUM = int(input("Ingrese un numero: "))
if NUM >= 0:
    R = "Positivo"
else:
   R = "Negativo"
print("El numero es: ", R)
print("Fin del programa")
print("Programa realizado por Daniel Gallegos NL = 20")
```
